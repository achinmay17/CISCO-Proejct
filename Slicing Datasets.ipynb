```python
case1 = pd.read_csv('C:\\Users\\ajnad\\Desktop\\Cisco\\Analysis\\Time series analysis\\forduke\\CaseE_part1.csv')
data = pd.read_csv('bgpclear_spine3_dataset.csv')

tmax = case1['time'].max()
tmin = case1['time'].min()


summary = pd.DataFrame()

data1 = data[(data['time'] >= tmin) & (data['time'] <= tmax)]

leaves = ['leaf1', 'leaf2', 'leaf3', 'leaf4','leaf5', 'leaf6', 'leaf7', 'leaf8','spine1','spine2','spine3','spine4']

for leaf in leaves:
    
    
    #Producer = 'leaf1'
    df = data1[data1['Producer']== leaf]

    #Encoding path 'data-rate'
    df = df[df['EncodingPath'] == 'Cisco-IOS-XR-infra-statsd-oper:infra-statistics/interfaces/interface/latest/data-rate']

    #Selecting columns
    df = df[['time','input-data-rate','input-packet-rate','output-data-rate','output-packet-rate']]

    df = df.apply(pd.to_numeric)

    #Plots
    df.plot(x = 'time',y = 'input-data-rate', title = leaf,figsize=(10,5))
    #df[1:5000].plot(x = 'time', y ='output-data-rate', title = leaf, figsize=(10,5))
    
    
    temp = pd.DataFrame(df.describe())
    temp = temp[['input-data-rate']]
    temp.columns = [str(leaf) + ' input data rate']
    temp = temp.drop(['count','std','min','25%','50%','75%'])
    
    summary = pd.concat([summary, temp],axis =1 )
```

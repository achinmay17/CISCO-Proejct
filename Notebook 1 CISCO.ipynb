```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
#Reading data frame bgpclear
data1 = pd.read_csv('bgpclear842017.csv')
```

```python
leaves = ['leaf1', 'leaf2', 'leaf3', 'leaf4','leaf5', 'leaf6', 'leaf7', 'leaf8','spine1','spine2','spine3','spine4']

for leaf in leaves:
    
    
    #Producer = 'leaf1'
    df = data1[data1['Producer']== leaf]

    #Encoding path 'data-rate'
    df = df[df['EncodingPath'] == 'Cisco-IOS-XR-infra-statsd-oper:infra-statistics/interfaces/interface/latest/data-rate']

    #Selecting columns
    df = df[['time','input-data-rate','input-packet-rate','output-data-rate','output-packet-rate']]

    df = df.apply(pd.to_numeric)

    #Plots
    df[1:5000].plot(x = 'time',y = 'input-data-rate', title = leaf,figsize=(10,5))
    df[1:5000].plot(x = 'time', y ='output-data-rate', title = leaf, figsize=(10,5))    
        
```
